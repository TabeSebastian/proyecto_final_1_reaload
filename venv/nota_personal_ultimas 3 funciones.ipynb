```python
import re
import pandas as pd

# Define la función para extraer la información
def extraer_informacion(cadena):
    resultado = None  # Inicialmente establecemos el resultado como nulo

    match = re.search(r"'posted': 'Posted [A-Za-z]+\s+\d+\s+(\d+).*'item_id': '(\d+)'.*'recommend': (\w+)", cadena)

    if match:
        resultado = {}  # Creamos un diccionario solo si se encuentra una coincidencia

        anio = match.group(1)  # Extrae el año
        item_id = match.group(2)  # Extrae 'item_id'
        recommend = match.group(3)  # Extrae 'recommend'

        # Convierte 'recommend' a un valor booleano (True o False)
        recommend = recommend.lower() == 'true'

        # Almacena los resultados en el diccionario
        resultado['Año'] = anio
        resultado['item_id'] = item_id
        resultado['recommend'] = recommend

    return resultado

# Nombres de las columnas a las que aplicar la función
columnas_a_procesar = ['review_1', 'review_2', 'review_3', 'review_4', 'review_5', 'review_6']

# Aplica la función a las columnas especificadas y almacena los resultados en una nueva columna
for columna in columnas_a_procesar:
    df_user_review[columna + '_resultados'] = df_user_review[columna].apply(extraer_informacion)

# Ahora df_user_review contendrá nuevas columnas que almacenan la información extraída en forma de diccionarios o valores nulos

# Combina todos los diccionarios en una lista
lista_resultados = []
for columna in columnas_a_procesar:
    lista_resultados.extend(df_user_review[columna + '_resultados'].tolist())

# Ahora 'lista_resultados' contendrá todos los diccionarios de resultados extraídos, o valores nulos en caso de no encontrar coincidencias
```

```python
# Crea un diccionario para almacenar los resultados por año
top_juegos_por_anio = {}

# Filtra los resultados que cumplen con 'recommend' igual a True
resultados_filtrados = [resultado for resultado in lista_resultados if resultado and resultado['recommend']]

# Agrupa los resultados por año
for resultado in resultados_filtrados:
    año = resultado['Año']
    item_id = resultado['item_id']

    if año not in top_juegos_por_anio:
        top_juegos_por_anio[año] = []

    top_juegos_por_anio[año].append(item_id)

# Obtiene el top 3 de juegos por año
top_3_por_anio = {}
for año, juegos in top_juegos_por_anio.items():
    top_3 = juegos[:3] if len(juegos) >= 3 else juegos
    top_3_por_anio[año] = top_3

print("Top 3 de juegos MÁS recomendados por usuarios por año:")
print(top_3_por_anio)
```

```python
# Convierte el diccionario en un DataFrame de pandas
df = pd.DataFrame(top_3_por_anio)

# Transponer el DataFrame para que los años sean índices y las columnas sean los juegos
df = df.transpose()

# Define las columnas para el DataFrame
columnas = ['Puesto 1', 'Puesto 2', 'Puesto 3']

# Renombra las columnas del DataFrame con las columnas definidas
df.columns = columnas

# Guarda el DataFrame en un archivo CSV
df.to_csv('top_juegos_por_anio.csv')
```

```python
# Crea un diccionario para almacenar los resultados por año
bottom_juegos_por_anio = {}

# Filtra los resultados que cumplen con 'recommend' igual a False
resultados_filtrados = [resultado for resultado in lista_resultados if resultado and not resultado['recommend']]

# Agrupa los resultados por año
for resultado in resultados_filtrados:
    año = resultado['Año']
    item_id = resultado['item_id']

    if año not in bottom_juegos_por_anio:
        bottom_juegos_por_anio[año] = []

    bottom_juegos_por_anio[año].append(item_id)

# Obtiene el top 3 de juegos menos recomendados por año
bottom_3_por_anio = {}
for año, juegos in bottom_juegos_por_anio.items():
    bottom_3 = juegos[:3] if len(juegos) >= 3 else juegos
    bottom_3_por_anio[año] = bottom_3

if '2010' in bottom_3_por_anio:
    del bottom_3_por_anio['2010']

print("Top 3 de juegos MENOS recomendados por usuarios por año:")
print(bottom_3_por_anio)
```

```python
# Convierte el diccionario en un DataFrame de pandas
df = pd.DataFrame(bottom_3_por_anio)

# Transponer el DataFrame para que los años sean índices y las columnas sean los juegos
df = df.transpose()

# Define las columnas para el DataFrame
columnas = ['Puesto 1', 'Puesto 2', 'Puesto 3']

# Renombra las columnas del DataFrame con las columnas definidas
df.columns = columnas

# Guarda el DataFrame en un archivo CSV
df.to_csv('bottom_3_por_anio.csv')
```

```python
# Crea un diccionario para almacenar las cuentas por año
cuentas_por_anio = {}

# Filtra los resultados que tienen 'recommend' igual a True o False
resultados_filtrados = [resultado for resultado in lista_resultados if resultado]

# Cuenta los registros según el análisis de sentimiento y el año
for resultado in resultados_filtrados:
    año = resultado['Año']
    recomendado = resultado['recommend']

    if año not in cuentas_por_anio:
        cuentas_por_anio[año] = {'Positive': 0, 'Negative': 0}

    if recomendado:
        cuentas_por_anio[año]['Positive'] += 1
    else:
        cuentas_por_anio[año]['Negative'] += 1

print("Cuenta de registros de reseñas de usuarios categorizados con un análisis de sentimiento según el año:")
print(cuentas_por_anio)
```

```python
# Convierte el diccionario en un DataFrame de pandas
df = pd.DataFrame(cuentas_por_anio)

# Transponer el DataFrame para que los años sean índices y las columnas sean los juegos
df = df.transpose()

# Define las columnas para el DataFrame
columnas = ['Positive', 'Negative']

# Renombra las columnas del DataFrame con las columnas definidas
df.columns = columnas

# Guarda el DataFrame en un archivo CSV
df.to_csv('def sentiment_analysis.csv')
```
