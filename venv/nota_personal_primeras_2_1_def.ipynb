```python
import re

# Inicializa un diccionario para realizar un seguimiento del tiempo de juego por juego
horas_por_juego = {}

# Itera a través de todas las filas
for index, row in df_user_item_clean.iterrows():
    for col in df_user_item_clean.columns[1:]:  # Comienza desde la segunda columna
        # Obtén los datos de la columna actual
        datos_fila = row[col]

        # Asegúrate de que los datos de la fila no sean None
        if datos_fila is not None:
            # Utiliza expresiones regulares para encontrar los valores
            match_item_id = re.search(r"'item_id': '(\d+)'", datos_fila)
            match_item_name = re.search(r"'item_name': '([^']+)'", datos_fila)
            match_playtime_forever = re.search(r"'playtime_forever': (\d+)", datos_fila)
            match_playtime_2weeks = re.search(r"'playtime_2weeks': (\d+)", datos_fila)

            if match_item_id:
                item_id = int(match_item_id.group(1))
                item_name = match_item_name.group(1) if match_item_name else None
                playtime_forever = int(match_playtime_forever.group(1)) if match_playtime_forever else 0
                playtime_2weeks = int(match_playtime_2weeks.group(1)) if match_playtime_2weeks else 0

                # Verifica si el juego ya está en el diccionario
                if item_id in horas_por_juego:
                    # Si el juego ya existe en el diccionario, suma las horas
                    horas_por_juego[item_id]['playtime_forever'] += playtime_forever
                    horas_por_juego[item_id]['playtime_2weeks'] += playtime_2weeks
                else:
                    # Si el juego no está en el diccionario, crea una entrada
                    horas_por_juego[item_id] = {
                        'item_name': item_name,
                        'playtime_forever': playtime_forever,
                        'playtime_2weeks': playtime_2weeks
                    }

# Convierte el diccionario en una lista de juegos con sus horas
lista_juegos_horas = [
    (item_id, juego['item_name'], juego['playtime_forever'], juego['playtime_2weeks'])
    for item_id, juego in horas_por_juego.items()
]

# Imprime la lista de juegos con sus horas
print(lista_juegos_horas)
```

```python
import pandas as pd

# Supongamos que tienes el DataFrame original llamado UserForGenre con las columnas user_id, item_id y playtime_forever

# Utiliza pivot_table para crear el nuevo DataFrame, agregando los valores duplicados
df_pivoted = pd.pivot_table(UserForGenre, index='item_id', columns='user_id', values='playtime_forever', aggfunc='sum')

# Rellenar los valores NaN (que representan user_id que no tenían playtime_forever) con 0
df_pivoted = df_pivoted.fillna(0)

# Reiniciar los nombres de las columnas y restablecer el índice
df_pivoted.columns.name = None
df_pivoted = df_pivoted.reset_index()

# Rename column 'item_id' to 'id'
df_pivoted = df_pivoted.rename(columns={'item_id': 'id'})

# Mostrar el DataFrame pivotado
df_pivoted.head()
```

```python
# Suponiendo que tienes los DataFrames df_pivoted y PlayTimeGenre
# Ambos tienen una columna común llamada "item_id"

# Unir los DataFrames por la columna "item_id" con PlayTimeGenre como el DataFrame principal
merged_df = PlayTimeGenre.merge(df_pivoted, on="id", how="inner")

# El parámetro "how" se establece en "inner" para que solo se incluyan las filas con "item_id" comunes en ambos DataFrames.
# Si deseas un tipo de unión diferente, puedes cambiar el valor de "how" a "left", "right", o "outer" según tus necesidades.

# Change column type to string for column: 'genres_0'
merged_df = merged_df.astype({'genres_0': 'string'})

# Mostrar el DataFrame resultante
merged_df.head()
```
